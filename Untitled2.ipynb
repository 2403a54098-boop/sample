# Task
Analyze patient treatment and outcome data from "/content/patients.csv", "/content/TREATMENT.csv", and "/content/OUTCOMES.csv". Merge the datasets, calculate average recovery days and success rate per treatment type, and pivot the data to show total cost per treatment type for each patient.

## Load data

### Subtask:
Load the three datasets (`patients.csv`, `TREATMENT.csv`, and `OUTCOMES.csv`) into pandas DataFrames.


**Reasoning**:
Import pandas and load the three CSV files into dataframes as instructed.



```python
import pandas as pd

df_patients = pd.read_csv('/content/patients.csv')
df_treatment = pd.read_csv('/content/TREATMENT.csv')
df_outcomes = pd.read_csv('/content/OUTCOMES.csv')
```

## Merge data

### Subtask:
Merge the three DataFrames into a single DataFrame based on the patient ID.


**Reasoning**:
Merge the three dataframes based on the patient ID and treatment ID as instructed.



```python
df_merged = pd.merge(df_patients, df_treatment, left_on='pateint_id', right_on='patient_id')
df_merged = pd.merge(df_merged, df_outcomes, left_on='treatment_id', right_on='TREATMENT_ID')
```

**Reasoning**:
The merging failed because of a KeyError, which indicates that the column 'patient_id' does not exist in one of the dataframes. I need to inspect the column names of the dataframes to identify the correct column names for merging.



```python
print(df_patients.columns)
print(df_treatment.columns)
print(df_outcomes.columns)
```

**Reasoning**:
The column names were inspected and it was found that `df_patients` has 'pateint_id' instead of 'patient_id'. The merge should be performed using the correct column names.



```python
df_merged = pd.merge(df_patients, df_treatment, left_on='pateint_id', right_on='patient_id')
df_merged = pd.merge(df_merged, df_outcomes, left_on='treatment_id', right_on='TREATMENT_ID')
```

**Reasoning**:
The second merge failed with a KeyError on 'TREATMENT_ID'. I need to check the columns of the dataframe resulting from the first merge and the `df_outcomes` dataframe to identify the correct column names for the second merge.



```python
print(df_merged.columns)
print(df_outcomes.columns)
```

**Reasoning**:
The column names were inspected and it was found that the intermediate merged dataframe has 'treatment_id' while `df_outcomes` has 'TREATMENT_ID'. The second merge should be performed using these correct column names.



```python
df_merged = pd.merge(df_merged, df_outcomes, left_on='treatment_id', right_on='TREATMENT_ID')
display(df_merged.head())
```

## Calculate average recovery days and success rate

### Subtask:
Group the merged DataFrame by `treatment_type` and calculate the average `recovery_days` and the success rate (proportion of `SUCCESS` that is 'YES').


**Reasoning**:
Group the merged DataFrame by 'treatment_type' and calculate the mean of 'RECOVERY_DAYS' and the success rate of 'SUCCESS'.



```python
treatment_analysis = df_merged.groupby('treatment_type').agg(
    average_recovery_days=('RECOVERY_DAYS', 'mean'),
    success_rate=('SUCCESS', lambda x: (x == 'YES').mean())
)
display(treatment_analysis)
```

## Pivot data

### Subtask:
Pivot the merged DataFrame to show each patient's total `cost` per `treatment_type`.


**Reasoning**:
Pivot the merged dataframe to show the total cost per treatment type for each patient and display the result.



```python
patient_treatment_cost = df_merged.pivot_table(
    index='pateint_id',
    columns='treatment_type',
    values='cost',
    aggfunc='sum'
)
display(patient_treatment_cost)
```

## Display results

### Subtask:
Display the results from the calculations and the pivoted table.


**Reasoning**:
Display the `treatment_analysis` and `patient_treatment_cost` DataFrames as requested by the instructions.



```python
display(treatment_analysis)
display(patient_treatment_cost)
```
