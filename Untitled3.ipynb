```python
import matplotlib.pyplot as plt

# Data
categories = ["Electronics", "Clothing", "Home Appliances", "Groceries"]
units_sold = [350, 220, 150, 500]

# Plot
plt.bar(categories, units_sold,color=['red','black','blue','green'])
plt.title("Monthly Sales by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Units Sold")
plt.show()
```
